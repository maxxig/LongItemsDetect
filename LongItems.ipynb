```python
df_attr = pd.read_csv('attr_for_all_products.csv', sep = ';')
```

```python
df_attr_process['attribute_value'] = df_attr_process.apply(lambda raw: raw['attribute_type_name'].strip() + ' equal ' + str(raw['value']).strip(), axis='columns')
df_attr_process['attr_concat_value'] = df_attr_process.groupby(['product_ID'])['attribute_value'].transform(lambda x: ' '.join(x.astype(str)))
```
